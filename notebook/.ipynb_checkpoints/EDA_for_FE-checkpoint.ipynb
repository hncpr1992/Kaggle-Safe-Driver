remove ps_calc* 

```python
col_ps_calc = train.columns[train.columns.str.startswith("ps_calc")]
```

```python
train_col_ps_calc = train[col_ps_calc]
```

```python
train = train.drop(col_ps_calc, axis=1)
```

```python
corr_mat = train.drop("id", axis=1).corr()

plt.figure(figsize=(10,12));
sns.heatmap(corr_mat, square=True);
```

```python
def binary_table(data, target, by):
    data.groupby(by)[target].value_counts().unstack().apply(lambda x: x/x.sum(), axis=1).plot.bar(stacked=True);
```

1.Reg

```python
binary_table(train, "target", "ps_reg_01");
```

```python
ps_reg_01_0_1 = train.groupby("ps_reg_01")["target"].value_counts().unstack()

ps_reg_01_0_1_ratio = ps_reg_01_0_1.sort_values(0, ascending=False).apply(lambda x: x/x.sum(), axis=1)

ps_reg_01_0_1_plus_ratio = ps_reg_01_0_1.merge(ps_reg_01_0_1_ratio, left_index=True, right_index=True)

ps_reg_01_0_1_plus_ratio
```

```python
binary_table(train, "target", "ps_reg_02");
```

```python
ps_reg_02_0_1 = train.groupby("ps_reg_02")["target"].value_counts().unstack()

ps_reg_02_0_1_ratio = ps_reg_02_0_1.sort_values(0, ascending=False).apply(lambda x: x/x.sum(), axis=1)

ps_reg_02_0_1_plus_ratio = ps_reg_02_0_1.merge(ps_reg_02_0_1_ratio, left_index=True, right_index=True)

ps_reg_02_0_1_plus_ratio
```

No need to encode

```python
ps_reg_03_0_1 = train.groupby("ps_reg_03")["target"].value_counts().unstack()

ps_reg_03_0_1_ratio = ps_reg_03_0_1.sort_values(0, ascending=False).apply(lambda x: x/x.sum(), axis=1)

ps_reg_03_0_1_plus_ratio = ps_reg_03_0_1.merge(ps_reg_03_0_1_ratio, left_index=True, right_index=True)

ps_reg_03_0_1_pos_index = ps_reg_03_0_1_plus_ratio.iloc[(ps_reg_03_0_1_plus_ratio["0_x"].values > 300) * (ps_reg_03_0_1_plus_ratio["1_y"].values > 0.04)].index

ps_reg_03_0_1_pos_map = pd.Series([1 if x in ps_reg_03_0_1_pos_index else 0 for x in train["ps_reg_03"]])
```
